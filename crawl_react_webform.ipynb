```python
import os
import pandas as pd
from bs4 import BeautifulSoup as bs, NavigableString, Comment
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

import os
import sys
module_path = os.path.abspath(os.path.join('./form_crawler'))
if module_path not in sys.path:
    sys.path.append(module_path)

from form_crawler.html_transformer import simplify_form
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.get('http://localhost:3000/antd')
```

```python
sources = []
targets = []
```

```python
DATA_COUNT = 500
```

```python
i = -1
while len(sources) < DATA_COUNT:
    try:
        i += 1
        driver.refresh()

        form = driver.find_element(By.ID, 'form')
        description = driver.find_element(By.ID, 'json-object').get_attribute('innerHTML')

        cleaned_form = simplify_form(driver, form)

        sources.append(cleaned_form)
        targets.append(description)

        if i % 5 == 0:
            print(i)
        
    except Exception as e:
        print(e, i)
        pass
```

```python
driver.close()
```

```python
df = pd.DataFrame({ "prompt": sources, "completion": targets })
```

```python
df.completion = df.completion.map(lambda x: " " + x)
```

```python
df = df[df.completion != " []"]
```
