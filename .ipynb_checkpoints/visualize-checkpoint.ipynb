```python
g = sns.catplot(
    data=df, kind="bar",
    x="method", y="value", hue="metric", height=5
)
# for i in g.containers:
#     g.bar_label(i,)
# g.despine(left=True)
g.set_axis_labels("method", "percentage")
g.legend.set_title("")
```

```python
# Separate the dataframe into 2 based on metric
df_metric1 = df[df['metric'] == 'FSS']
df_metric2 = df[df['metric'] == 'PR']

plt.figure(figsize=(15, 8))

# plot metric1
plt.subplot(1, 2, 1)
barplot1 = sns.barplot(x='method', y='value', data=df_metric1)
plt.title('FSS')
plt.ylim(0, df['value'].max() + 0.2 * df['value'].max())  # add some space for annotations

# Annotating the bars with the metric values
for p in barplot1.patches:
    barplot1.annotate(format(p.get_height(), '.2f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center',
                      va = 'center',
                      xytext = (0, 10),
                      textcoords = 'offset points')

# plot metric2
plt.subplot(1, 2, 2)
barplot2 = sns.barplot(x='method', y='value', data=df_metric2)
plt.title('PR')
plt.ylim(0, df['value'].max() + 0.2 * df['value'].max())  # add some space for annotations

# Annotating the bars with the metric values
for p in barplot2.patches:
    barplot2.annotate(format(p.get_height(), '.2f'),
                      (p.get_x() + p.get_width() / 2., p.get_height()),
                      ha = 'center',
                      va = 'center',
                      xytext = (0, 10),
                      textcoords = 'offset points')

plt.tight_layout()
plt.show()
```

```python
# pivot the data
df_pivot = df.pivot(index='method', columns='metric', values='value')

# plotting
plt.figure(figsize=(10, 6))
barplot = df_pivot.plot(kind='bar')

plt.title('Metric Comparison')
plt.ylabel('Value')
plt.ylim(0, df['value'].max() + 0.2 * df['value'].max())  # add some space for annotations

# Annotating the bars with the metric values
for rect in barplot.patches:
    # Find where everything is located
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()

    # The height of the bar is the data value and can be used as the label
    label_text = f'{height:.2f}'

    # ax.text(x, y, text)
    label_x = x + width / 2
    label_y = y + height + 0.01  # adjust 0.01 depending on the distance you want the annotation to have from the bars

    # plot only when height is greater than specified value
    if height > 0:
        barplot.text(label_x, label_y, label_text, ha='center', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
